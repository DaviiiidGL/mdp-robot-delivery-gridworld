### 2.2 Propiedad de Markov
La propiedad de Markov establece que el futuro depende del estado presente, no de toda la historia pasada:
\[
\mathbb{P}(X_{t+1} \mid X_t, X_{t-1}, \dots, X_0) = \mathbb{P}(X_{t+1} \mid X_t)
\]

### 2.4 Proceso de Decisión de Markov (MDP)
Un Proceso de Decisión de Markov extiende una cadena de Markov al incluir acciones elegidas por un agente. Un MDP finito descontado se define típicamente como:
\[
\mathcal{M} = (S, A, P, R, \gamma)
\]
donde:
- \( S \): espacio de estados
- \( A \): espacio de acciones
- \( P(s' \mid s,a) \): probabilidad de transición
- \( R(s,a,s') \): función de recompensa
- \( \gamma \in [0,1) \): factor de descuento
